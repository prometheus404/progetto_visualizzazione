```python
# Livello regionale

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

inqui = pd.read_csv('../csv/fumatori+tumori.csv', encoding='utf-8', sep=',')
inqui_s = inqui.sort_values(by=["Fumatori + ex (/1000)"], ascending=True)
inqui_s.plot.scatter(x='Morti tumori', y='Fumatori + ex (/1000)')

plt.show()

# Questo dimostra che più gente fuma in una regione
# più morti di tumore (vie respiratorie) ha quella regione.
```

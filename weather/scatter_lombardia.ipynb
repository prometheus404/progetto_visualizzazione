```python
import pandas as pd, plotly.express as px, matplotlib.pyplot as plt, numpy as np, scipy
from scipy import stats
```

```python
#Pollutants lists
PM10 = ['PM10', 'PM10 (SM2005)']
NO2 = ['Biossido di Azoto']
PM25 = ['Particelle sospese PM2.5']
CO_8h = ['Monossido di Carbonio']
O3 = ['Ozono']
SO2 = ['Biossido di Zolfo']
C6H6 = ['Benzene']

#Weather attributes lists
HUM = ['Umidità Relativa']
WIND_DIR = ['Direzione Vento']
TEMP = ['Temperatura']
WIND_SPEED = ['Velocità Vento']
SNOW = ['Altezza Neve']
RADIATIONS = ['Radiazione Globale']
RAIN = ['Precipitazione']

#Years available
years = range(2016, 2019)
```

```python
def weather_pollutant(year, pollutant = [], weather_attribute = []):
    csv = pd.read_csv('../csv/lombardia/scatter/'+str(year)+'/mean_'+str(year)+'_'+pollutant[0]+'_'+weather_attribute[0]+'.csv', encoding='utf-8', sep=',')
    #Remove outliers
    if weather_attribute[0]=='Precipitazione': 
        csv.drop(csv.index[(csv["Valore meteo"] > 0.4 )],axis=0,inplace=True)
    #Plot
    csv.plot.scatter(x='Valore meteo', y='Valore inquinante')
    plt.xlabel('Weather attribute: '+weather_attribute[0])
    plt.ylabel('Pollutant: '+pollutant[0])
    plt.show()
    weather = np.array(csv['Valore meteo'])
    pollutant = np.array(csv['Valore inquinante'])
    #Pearson
    print(scipy.stats.pearsonr(weather, pollutant))
    #Spearman
    print(scipy.stats.spearmanr(weather, pollutant))
    #Kendall
    print(scipy.stats.kendalltau(weather, pollutant))
```

```python
weather_pollutant(2016, PM10, HUM)
weather_pollutant(2017, PM10, HUM)
weather_pollutant(2018, PM10, HUM)
weather_pollutant(2016, PM10, TEMP)
weather_pollutant(2017, PM10, TEMP)
weather_pollutant(2018, PM10, TEMP)
weather_pollutant(2016, PM10, WIND_SPEED)
weather_pollutant(2017, PM10, WIND_SPEED)
weather_pollutant(2018, PM10, WIND_SPEED)
weather_pollutant(2016, PM10, RADIATIONS)
weather_pollutant(2017, PM10, RADIATIONS)
weather_pollutant(2018, PM10, RADIATIONS)
weather_pollutant(2016, PM10, RAIN)
weather_pollutant(2017, PM10, RAIN)
weather_pollutant(2018, PM10, RAIN)
```
